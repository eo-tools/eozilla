From https://airflow.apache.org/docs/apache-airflow/stable/security/api.html:

Get API access token (JWT):
```
curl -X POST ${ENDPOINT_URL}/auth/token \
  -H "Content-Type: application/json" \
  -d '{
    "username": "your-username",
    "password": "your-password"
  }'
```

Test API call:
```
curl -X GET ${ENDPOINT_URL}/api/v2/dags \
  -H "Authorization: Bearer <JWT-TOKEN>"
```

````python
import requests
````

````python
ENDPOINT_URL = "http://localhost:8080"
````

````python
response = requests.post(
    f"{ENDPOINT_URL}/auth/token", 
    headers={"Content-Type": "application/json"},
    json={"username": "admin", "password": "xxxxxxxxxx"}, 
)
response
````

````python
access_token_result = response.json()
````

````python
response = requests.get(
    f"{ENDPOINT_URL}/api/v2/dags", 
    headers={
        "Content-Type": "application/json",
        "Authorization": f"Bearer {access_token_result['access_token']}",
    },    
)
response
````
